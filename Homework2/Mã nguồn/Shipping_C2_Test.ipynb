```python
def calculateShipping_fee(weight, distance):
    if weight < 0.1 or weight > 50.0 or distance < 1 or distance > 2000:
        return "Đầu vào không hợp lệ"

    if weight <= 1.0:
        base_fee = 15000
    elif weight <= 5.0:
        base_fee = 25000
    elif weight <= 20.0:
        base_fee = 45000
    else:
        base_fee = 80000

    if distance <= 100:
        factor = 1.0
    elif distance <= 500:
        factor = 1.5
    else:
        factor = 2.5

    total = base_fee * factor

    if weight > 30.0 and distance > 500:
        total += 20000

    return total
```

```python
test_cases = [
    {"id": 1, "weight": 60.0, "distance": 100, "expected": "Đầu vào không hợp lệ"},
    {"id": 2, "weight": 0.5, "distance": 50, "expected": 15000.0},
    {"id": 3, "weight": 3.0, "distance": 300, "expected": 37500.0},
    {"id": 4, "weight": 15.0, "distance": 600, "expected": 112500.0},
    {"id": 5, "weight": 40.0, "distance": 1000, "expected": 220000.0}
]
```

```python
print(f"{'ID':<5} | {'Input (w, d)':<15} | {'Expected':<20} | {'Actual':<20} | {'Result':<10}")
print("-" * 80)

for tc in test_cases:
    actual = calculateShipping_fee(tc["weight"], tc["distance"])
    
    result = "PASS" if actual == tc["expected"] else "FAIL"
    
    input_str = f"({tc['weight']}, {tc['distance']})"
    
    print(f"{tc['id']:<5} | {input_str:<15} | {str(tc['expected']):<20} | {str(actual):<20} | {result:<10}")
```
